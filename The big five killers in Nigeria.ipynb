```python
# Calculate the total top five causes of death in a decade
n2 = Top_five_killers_df.copy()
n2['Top_five'] = (n2['Diarrheal Diseases'] +
            n2['Malaria'] +
            n2['Lower Respiratory Infections'] +
            n2['Neonatal Disorders'] +
            n2['Cardiovascular Diseases'])
n2
```

```python
# Total sum of all causes of death in Nigeria
import pandas as pd

N = Nigeria_killers_sorted.drop(columns=['Year', 'Country', 'Code'])

# Calculate row sums
Total2 = []
for _, row in N.iterrows():
    row_sum = row.sum()
    Total2.append(row_sum)

# Assign the results to a new column in Top_five_killers_df_copy
n2['Total'] = Total2[:len(n2)]

# Display the modified DataFrame
#print(Top_five_killers_df)
n2
```

```python
# Calculate the proportion of Top five causes of death over a decade in Nigeria
n2['Diarrheal_prop'] = n2['Diarrheal Diseases'] / n2['Total']*100
n2['Malaria_prop'] = n2['Malaria'] / n2['Total']*100
n2['Lower Respiratory Infections_prop'] = n2['Lower Respiratory Infections'] / n2['Total']*100
n2['Neonatal Disorders_prop'] = n2['Neonatal Disorders'] / n2['Total']*100
n2['Cardiovascular Diseases_prop'] = n2['Cardiovascular Diseases'] / n2['Total']*100

n2['others'] = n2['Total'] - n2['Top_five']
n2['others_prop'] = n2['others']/n2['Total']*100
n2
```

```python
# plot a stacked horizontal bar chart
# Data for the stacked horizontal bar chart
Years = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
proportions1 = n2['Diarrheal_prop']
proportions2 = n2['Malaria_prop']
proportions3 = n2['Lower Respiratory Infections_prop']
proportions4 = n2['Neonatal Disorders_prop']
proportions5 = n2['Cardiovascular Diseases_prop']
proportions6 = n2['others_prop']

# Plotting the stacked horizontal bar chart
plt.figure(figsize=(8, 6))

bar_positions = np.arange(len(Years))

bar1 = plt.barh(bar_positions, proportions1, label='Diarrheal')
bar2 = plt.barh(bar_positions, proportions2, left=proportions1, label='Malaria')
bar3 = plt.barh(bar_positions, proportions3, left=proportions1 + proportions2,
                 label='Lower Respiratory Infections')
bar4 = plt.barh(bar_positions, proportions4, left=proportions1 + proportions2 + proportions3,
                 label='Neonatal Disorders')
bar5 = plt.barh(bar_positions, proportions5,
                 left=proportions1 + proportions2 + proportions3 + proportions4,
                 label='Cardiovascular Diseases')
bar6 = plt.barh(bar_positions, proportions6,
                 left=proportions1 + proportions2 + proportions3 + proportions4 + proportions5,
                 label='Others')

# Adding data point labels
for bars in [bar1, bar2, bar3, bar4, bar5, bar6]:
    for bar in bars:
        width = bar.get_width()
        x_pos = bar.get_x() + width / 2
        if width < 2:
            plt.text(x_pos, bar.get_y() + bar.get_height() / 2, f'{width:.1f}%', ha='center', va='center', fontsize=9,
                     bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round'))
        else:
            plt.text(x_pos, bar.get_y() + bar.get_height() / 2, f'{width:.1f}%', ha='center', va='center',
                     fontsize=9, color='black',
                     bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round'))

# Adding labels and title
plt.xlabel('Proportion (%)')
plt.ylabel('Years')
plt.title('Percentage of Causes of Death in a Decade')
plt.yticks(bar_positions, Years)
plt.legend()

# Displaying the chart
plt.show()
```
